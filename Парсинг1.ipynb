```python
#Взяли депутатов Свердловской области
import requests
url = "http://zsso.ru/structure/deputies/"
page = requests.get(url)
page
```

```python
#используем BeautifulSoup, чтоб смотреть необходимые нам данные
from bs4 import BeautifulSoup
text = page.text
soup = BeautifulSoup(text, "lxml")
soup
```

```python
print (soup.prettify()) #для того чтобы было удобно смотреть используем prettify
```

```python
soup.find_all('div', {'class':'section-list type-3 margin'})#выбираем только те данные в которых есть имена и фракции
```

```python
len(soup.find_all('div', {'class':'section-list type-3 margin'}))
```

```python
table = soup.find_all('div', {'class':'section-list type-3 margin'})[0]
```

```python
for t in table.find_all('div', {'class':'sl-item-name hh2'}):
    print(t.text) #выбираем имена
```

```python
names = []
for t in table.find_all('div', {'class':'sl-item-name hh2'}):
    names.append(t.text) 
```

```python
names
```

```python
fracs = []
for t in table.find_all('div', {'class':'sl-item-fraction'}):
    fracs.append(t.text.strip()) #выбираем фракции
```

```python
fracs
```

```python
len(fracs)
```

```python
len(names)
```

```python
#количество имен и фракций одинаковые, значит все нормально
```

```python
import pandas as pd
df = pd.DataFrame({'name': names, 'party': fracs})
df.head() #создаем таблицу
```
