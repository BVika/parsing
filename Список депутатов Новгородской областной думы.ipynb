```python
#Список депутатов Новгородской областной думы
import requests
url = "https://www.novoblduma.ru/duma/deputies/"
page = requests.get(url)
page
```

```python
#используем BeautifulSoup, чтобы увидеть нужные данные
from bs4 import BeautifulSoup
text = page.text
soup = BeautifulSoup(text, "lxml")
soup
```

```python
print (soup.prettify()) 
```

```python
soup.find_all('div', {'class':'deputy_page-list'})#
```

```python
table = soup.find_all('div', {'class':'deputy_page-list'})[0]
```

```python
for t in table.find_all('div', {'class':'deputy-name'}):
    print(t.text) #выбираем имена
```

```python
names = []
for t in table.find_all('div', {'class':'deputy-name'}):
    names.append(t.text) 
names
```

```python
party = []
for t in table.find_all('div', {'class':'deputy-party'}):
    party.append(t.text.strip()) 
    
party
```

```python
print(len(names))
print(len(party))
```

```python
import pandas as pd
df = pd.DataFrame({'name': names, 'party': party})
df.head() 
```
